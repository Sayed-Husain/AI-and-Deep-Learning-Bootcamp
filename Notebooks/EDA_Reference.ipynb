## 1. Introduction to Pandas

### 1.1 What is Pandas?

Pandas is a powerful open-source data manipulation and analysis library for Python. It provides data structures and functions that make it easy to work with structured data.

### 1.2 Key Data Structures: Series and DataFrame

- Series: A one-dimensional labeled array that can hold data of various types.
- DataFrame: A two-dimensional labeled data structure with columns of potentially different types.



| Series | DataFrame|
|:---: | :---:|
|**1 Dimensional** | **2 Dimensional**|
|Less Common to use | More common to use|
|Created using `pd.Series()` | Created using `pd.DataFrame()`|
|Based On Numpy Arrays | Based on Numpy Array|
|In some cases, swapable with Python's list `[]` | Lists within Dictionary `{column_name : []}`|



### 1.3 Loading Data

Pandas allows you to read various types of data formats, such as CSV, Excel, SQL, and more.

```python
import pandas as pd

# Reading a CSV file
data = pd.read_csv('data.csv')

# Reading an Excel file
data = pd.read_excel('data.xlsx')

# Reading from a SQL database
import sqlite3
conn = sqlite3.connect('database.db')
data = pd.read_sql_query('SELECT * FROM table', conn)
```

### 1.4 Data Manipulation
Pandas provides powerful tools for data manipulation, including filtering, grouping, merging, and more.


```python
# Filtering data
filtered_data = data[data['column'] > 10]

# Grouping and aggregating data
grouped_data = data.groupby('category')['value'].mean()

# Merging data
merged_data = pd.merge(data1, data2, on='key_column')

# Adding a new column
data['new_column'] = data['column1'] + data['column2']

```


## 2. Introduction to NumPy
### 2.1 What is NumPy?
NumPy is the fundamental package for scientific computing with Python. It provides support for large, multi-dimensional arrays and matrices, as well as various mathematical functions to operate on these arrays.

### 2.2 NumPy Arrays (ndarrays)
NumPy arrays, or ndarrays, are the core data structure in NumPy. They are similar to Python lists but optimized for numerical operations.

### 2.3 Array Creation and Manipulation

```python
import numpy as np

# Creating arrays
arr = np.array([1, 2, 3, 4, 5])

# Multi-dimensional arrays
matrix = np.array([[1, 2, 3], [4, 5, 6]])

# Array operations
result = arr * 2
```

### 2.4 Mathematical Operations
NumPy provides a wide range of mathematical functions and operations on arrays.

```python
# Element-wise operations
sum_array = arr + arr
squared_array = arr ** 2

# Linear algebra
dot_product = np.dot(matrix1, matrix2)
inverse_matrix = np.linalg.inv(matrix)
```


## 3. Introduction to Matplotlib
### 3.1 What is Matplotlib?
Matplotlib is a popular data visualization library for creating static, animated, and interactive visualizations in Python.

### 3.2 Basic Plotting

```python
import matplotlib.pyplot as plt

# Line plot
plt.plot(x_values, y_values, label='line')

# Scatter plot
plt.scatter(x_values, y_values, label='points')

plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.legend()
plt.show()
```

### 3.3 Customizing Plots
Matplotlib provides options to customize various aspects of your plots.


```python
plt.plot(x, y, color='red', linestyle='dashed', marker='o', markersize=8, label='customized line')
```

### 3.4 Subplots and Multiple Plots
You can create subplots and multiple plots within a single figure.

```python
plt.subplot(2, 1, 1)  # 2 rows, 1 column, first subplot
plt.plot(x1, y1)

plt.subplot(2, 1, 2)  # 2 rows, 1 column, second subplot
plt.plot(x2, y2)

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

## Practical Example

````python
import pandas as pd

# Create a DataFrame about programming languages
data = {
    'Language': ['Python', 'Java', 'JavaScript', 'C++', 'Ruby', 'R'],
    'YearCreated': [1991, 1995, 1995, 1983, 1995, 1993],
    'Paradigm': ['Multi-paradigm', 'Object-oriented', 'Multi-paradigm', 'Multi-paradigm', 'Multi-paradigm', 'Multi-paradigm'],
    'PopularityRank': [1, 2, 3, 4, 15, 18]
}

programming_languages_df = pd.DataFrame(data)
````

### Basic DataFrame operations

````python
# Get the first few rows
print("\nFirst 3 rows:")
programming_languages_df.head(3)
````

````python
# Get summary statistics
print("Summary statistics:")
programming_languages_df.describe()
````

````python
# Filter languages created after 1990
recent_languages = programming_languages_df[programming_languages_df['YearCreated'] > 1990]
print("Languages created after 1990:")
recent_languages
````

````python
# Group languages by paradigm and calculate mean popularity rank
paradigm_groups = programming_languages_df.groupby('Paradigm')['PopularityRank'].mean()
print("Average popularity rank by paradigm:")
paradigm_groups
````

````python
# Sorting by popularity rank
sorted_by_popularity = programming_languages_df.sort_values(by='PopularityRank')
print("Languages sorted by popularity rank:")
sorted_by_popularity
````

````python
# Adding a new column
programming_languages_df['Age'] = 2023 - programming_languages_df['YearCreated']
print("Updated DataFrame with 'Age' column:")
programming_languages_df
````

````python
import numpy as np

# Extract popularity ranks as a NumPy array
popularity_ranks = programming_languages_df['PopularityRank'].values

# Calculate the mean and standard deviation of popularity ranks using NumPy
mean_rank = np.mean(popularity_ranks)
std_deviation = np.std(popularity_ranks)

print("Mean Popularity Rank:", mean_rank)
print("Standard Deviation of Popularity Ranks:", std_deviation)
````

````python
import matplotlib.pyplot as plt

# Create a bar plot using Matplotlib based on the age of programming languages
plt.bar(programming_languages_df['Language'], programming_languages_df['Age'])
plt.xlabel('Programming Languages')
plt.ylabel('Age')
plt.title('Age of Programming Languages')
plt.xticks(rotation=45) # rotate the name of the programming language
plt.show()
````
