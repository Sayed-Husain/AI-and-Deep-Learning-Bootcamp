# Covid-19 Notebook
> In this Notebook, we are going to explore the covid-19 dataset and learn the main functionalities of the Pandas, NumPy, and Matplotlib libraries.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the COVID-19 dataset
url = 'https://raw.githubusercontent.com/Sayed-Husain/AI-and-Deep-Learning-Bootcamp/main/Data/country_wise_latest.csv'  # Replace with the actual URL or file path
df = pd.read_csv(url)

# Data Exploration
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Display basic statistics using NumPy
confirmed_mean = np.mean(df['Confirmed'])
deaths_median = np.median(df['Deaths'])
recovered_max = np.max(df['Recovered'])

print(f"Mean Confirmed Cases: {confirmed_mean}")
print(f"Median Deaths: {deaths_median}")
print(f"Maximum Recovered Cases: {recovered_max}")
```

```python
# Bar Plot: Top 10 Countries by Confirmed Cases

plt.figure(figsize=(10, 6))
top_countries = df.nlargest(10, 'Confirmed')
plt.bar(top_countries['Country/Region'], top_countries['Confirmed'])
plt.xlabel('Country')
plt.ylabel('Confirmed Cases')
plt.title('Top 10 Countries by Confirmed Cases')
plt.xticks(rotation=45)
```

```python
# Line Plot: Trend of Confirmed Cases over Time
plt.figure(figsize=(10, 6))
plt.plot(df['Confirmed last week'], label='Last Week')
plt.plot(df['Confirmed'], label='Current Week')
plt.xlabel('Week')
plt.ylabel('Confirmed Cases')
plt.title('Trend of Confirmed Cases over Time')
plt.legend()
```

```python
# Scatter Plot: Confirmed vs. Deaths
plt.figure(figsize=(10, 6))
plt.scatter(df['Confirmed'], df['Deaths'])
plt.xlabel('Confirmed Cases')
plt.ylabel('Deaths')
plt.title('Confirmed vs. Deaths')
```

```python
# Region Specefic EDA
Region = df.groupby(by='WHO Region').sum()
```

```python
# Create a pie chart with enhanced settings
plt.figure(figsize=(10, 5))
Region['Recovered'].plot(kind='pie',  figsize=(10,5), labels=None)


# Add title and legend
plt.title('Recovered Cases Distribution by WHO Region')
plt.legend(labels=Region['Recovered'].index, loc='upper left')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal');
```

```python
# Create a pie chart with enhanced settings
plt.figure(figsize=(10, 5))
Region['Active'].plot(kind='pie',  figsize=(10,5), labels=None)


# Add title and legend
plt.title('Active Cases Distribution by WHO Region')
plt.legend(labels=Region['Recovered'].index, loc='upper left')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal');
```
