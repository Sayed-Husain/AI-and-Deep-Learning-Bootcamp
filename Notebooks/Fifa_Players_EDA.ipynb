## Best Player in the world

> Let's write code to figure out who are the best players in the world


For this experiment, we do not need the entire dataset, we only need the following columns:

* **Known As** - The player name
* **Overall** - The player's overall score
* **Potential** - The players potential

Therefore, we will create a new DataFrame (best_players) only containing these columns

```python
# Create the best_players DataFrame only containing the columns mentioned.
best_players = pd.DataFrame(df, columns=["Known As", "Overall", "Potential"])
```

```python
# View the first 5 rows of the data
best_players.head()
```

Though it seems that the DataFrame is already sorted, let's gurantee it by sorting it.

Though we achieved our intended results and figured out who are the best players in the world, looking at a table is not fun. Threfore, lets take an extra step to visualize the data we gathered.

## Highest and Lowest BMI

For this experiment, the columns we need to get are: **`Known As`** and the **`BMI`**.

However, the original dataset does not have a `BMI` coulmn, therefore, we are going to make our own using the height_cm and the weight_kg columns.

**`BMI`** = `weight(kg)`/`height(m)`^2

Therefore, we will make a new DataFrame **`players_BMI`** containing the following columns from the original dataset:

* `Known As`
* `Height(in cm)`
* `Weight(in kg)`

```python
# Create the players_BMI DataFrame containing the columns mentioned
players_BMI = pd.DataFrame(df, columns=["Known As", "Height(in cm)", "Weight(in kg)"])

# View the new DataFrame
players_BMI.head()
```

We will create a BMI column by deviding the **`Weight(in kg)`** by the **square of** `Height(in cm)	`/ **100**

```python
# Create a BMI column
players_BMI["BMI"] = players_BMI["Weight(in kg)"] / np.square(players_BMI["Height(in cm)"] / 100)

# View the head of the DataFrame
players_BMI.head()
```

Looking great so far.

Now let's do two things:

Find out who has the highest and lowest BMI in the game
Get the highest BMI, the Lowest BMI, and the median BMI and plot them.

```python
# Find the who has the highest and lowest BMI by sorting the DataFrame
players_BMI.sort_values("BMI", ascending=False)
```

```python
# Find the max, min, and median values
max_BMI = np.max(players_BMI["BMI"])
median_BMI = np.median(players_BMI["BMI"])
min_BMI = np.min(players_BMI["BMI"])
```

```python
# Plot the data
plt.figure(figsize=(10,5))
plt.barh(["Max", "Median", "Min"], [max_BMI, median_BMI, min_BMI])
plt.title("BMI in the Football Game")
```

## Find Best Investment

```python
# Create the players_investment DataFrame
players_investment = pd.DataFrame(df, columns=["Known As", "Value(in Euro)", "Wage(in Euro)"])

# Add value for wage
players_investment['value_wage_difference'] = players_investment['Value(in Euro)'] - players_investment['Wage(in Euro)']


# Add Bahraini currency to the DataFrame
players_investment["value_wage_difference_bd"] = players_investment["value_wage_difference"] * 0.38

# Sort the DataFrame
players_investment = players_investment.sort_values("value_wage_difference", ascending=False)

# View the head of the DataFrame
players_investment.head()
```

```python
# Visulaize the best 10 players to invest in
X = players_investment.head(10)["Known As"]
Y = players_investment.head(10)["value_wage_difference_bd"]

plt.figure(figsize=(20,7))
plt.bar(X, Y)
plt.title("Best Investment for 2022")
plt.xlabel("Player Name");
```

## Categorical EDA

### Create Histograms

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.hist(df['Age'], bins=20, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

plt.subplot(1, 3, 2)
plt.hist(df['Height(in cm)'], bins=20, edgecolor='black')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.title('Height Distribution')

plt.subplot(1, 3, 3)
plt.hist(df['Weight(in kg)'], bins=20, edgecolor='black')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.title('Weight Distribution')

plt.tight_layout()
plt.show()
```

### Pie Chart

```python
preferred_foot_counts = df['Preferred Foot'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(preferred_foot_counts, labels=preferred_foot_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Preferred Foot Distribution')
plt.axis('equal')
plt.show()
```
