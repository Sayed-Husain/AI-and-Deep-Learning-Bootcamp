# Neural Network Classification with TensorFlow


## Introduction to Classification with TensorFlow
Classification is a fundamental task in machine learning and deep learning, involving the categorization of data into distinct classes or labels. TensorFlow offers a powerful platform for building and training classification models. In this notebook, we'll dive into neural network classification using TensorFlow, starting with a simple linear model and gradually introducing non-linearity to improve model performance.

Here are some important notes to consider about classification in TensorFlow:

<br>

#### Loss Functions:
For binary classification, the Binary Cross-Entropy loss function is commonly used. For multi-class classification, the Categorical Cross-Entropy loss is typically employed. These loss functions help measure the difference between predicted and actual labels.


<br>

#### Activation Functions:
Activation functions introduce non-linearity to neural networks. For binary classification, the **`sigmoid`** activation function is often used in the output layer to produce a probability value between 0 and 1. For multi-class classification, the **`softmax`** activation function is employed to produce probability scores for each class, ensuring they sum up to 1.

<br>

#### Metrics:
In addition to the loss function, **`accuracy`** is a commonly used metric for classification tasks. However, depending on the problem, other metrics like **`precision`**, **`recall`**, and **`F1-score`** might be more appropriate.

<br>

## 1. Creating the Dataset
We'll begin by generating a synthetic dataset using the make_circles function from scikit-learn. This dataset will contain two classes of data points arranged in concentric circles.

```python
from sklearn.datasets import make_circles

# Make 1000 examples
n_samples = 1000

# Create circles
X, y = make_circles(n_samples, noise=0.03, random_state=42)

# Visualize the dataset
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Synthetic Classification Dataset")
plt.show()
```

## 2. Building and Compiling a Linear Model
Let's start by building a linear model, which is a neural network with a single neuron and no activation function. We'll observe its performance and then introduce non-linearity to improve it.

```python
# import tensorflow
import tensorflow as tf

# Create a linear model using the Sequential API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)  # Linear activation by default
])

# Compile the linear model
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                     optimizer=tf.keras.optimizers.SGD(),
                     metrics=['accuracy'])
```

```python
# Train the model
model.fit(X, y, epochs=5)
```

```python
# Evaluate our model
model.evaluate(X, y)
```

On a binary classification problem, 50% is the equivalent of guessing, however, our model is doing a worse job than guessing at this point.



## 3. Try Different Techniques to Improve the model

### 3.2 Increase Model Complexity

```python
# import tensorflow
import tensorflow as tf

# Create a linear model using the Sequential API
model = tf.keras.Sequential([
  tf.keras.layers.Dense(100), # add 100 dense neurons
  tf.keras.layers.Dense(10), # add another layer with 10 neurons
  tf.keras.layers.Dense(1)
])

# Compile the linear model
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                     optimizer=tf.keras.optimizers.SGD(),
                     metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=5)
```

```python
# Evaluate our model
model.evaluate(X, y)
```

No Observable improvement can be seen by increasing the model complexity. Since these two techniques (and other techniques) are not improving the model, there must be an issue with the model itself.

## 4. Introducing Non-Linearity: Adding Activation Function





### 4.1 Understand Non-lineraity

Before introducing Non-Linearity and adding Activation Function to our model, let's first understand how the current model is working

```python
import numpy as np

def plot_decision_boundary(model, X, y):
  """
  Plots the decision boundary created by a model predicting on X.
  This function has been adapted from two phenomenal resources:
   1. CS231n - https://cs231n.github.io/neural-networks-case-study/
   2. Made with ML basics - https://github.com/GokuMohandas/MadeWithML/blob/main/notebooks/08_Neural_Networks.ipynb
  """
  # Define the axis boundaries of the plot and create a meshgrid
  x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
  y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
  xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                       np.linspace(y_min, y_max, 100))

  # Create X values (we're going to predict on all of these)
  x_in = np.c_[xx.ravel(), yy.ravel()] # stack 2D arrays together: https://numpy.org/devdocs/reference/generated/numpy.c_.html

  # Make predictions using the trained model
  y_pred = model.predict(x_in)

  # Check for multi-class
  if model.output_shape[-1] > 1: # checks the final dimension of the model's output shape, if this is > (greater than) 1, it's multi-class
    print("doing multiclass classification...")
    # We have to reshape our predictions to get them ready for plotting
    y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
  else:
    print("doing binary classifcation...")
    y_pred = np.round(np.max(y_pred, axis=1)).reshape(xx.shape)

  # Plot decision boundary
  plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
  plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
```

```python
# Check out the predictions our model is making
plot_decision_boundary(model, X, y)
```

### 4.2 Bulid Non-Linear Model
Now, let's introduce non-linearity by adding an activation function to the model. We'll use the **`sigmoid`** activation for output of the binary classification.

```python
# Create a model with sigmoid activation using the Sequential API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation='relu'),  # relu activation
    tf.keras.layers.Dense(10, activation='relu'),  # relu activation
    tf.keras.layers.Dense(1, activation='sigmoid')  # Sigmoid activation
])

# Compile the sigmoid model
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=100)
```

```python
# Evaluate our model
model.evaluate(X, y)
```

```python
# Check out the predictions our model is making
plot_decision_boundary(model, X, y)
```
